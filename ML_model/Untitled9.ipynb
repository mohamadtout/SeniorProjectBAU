```python
import pandas as pd
from surprise import Dataset, Reader, SVD, dump
from surprise.model_selection import train_test_split

# Create synthetic data
data = {
    'user_id': [1, 2, 1, 3, 4, 2, 5, 3, 5, 1],
    'attraction_id': [10, 10, 20, 30, 20, 30, 10, 20, 30, 20],
    'rating': [5, 4, 5, 3, 4, 5, 3, 4, 5, 3]
}
df = pd.DataFrame(data)

# Load dataset into Surprise
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(df[['user_id', 'attraction_id', 'rating']], reader)

# Train-test split
trainset, testset = train_test_split(data, test_size=0.25)

# Use SVD algorithm
algo = SVD()

# Train the model
algo.fit(trainset)

# Optionally evaluate the model
predictions = algo.test(testset)

# Save the trained model
dump.dump('./your_trained_model.pkl', algo=algo)
```

```python
from flask import Flask, request, jsonify
from surprise import dump

app = Flask(__name__)

# Load your trained model
_, algo = dump.load('./your_trained_model.pkl')

@app.route('/recommend', methods=['POST'])
def recommend():
    data = request.get_json()
    user_id = data['user_id']
    attraction_id = data['attraction_id']

    # Generate prediction for this user and attraction
    prediction = algo.predict(user_id, attraction_id)

    return jsonify({'estimated_rating': prediction.est})

if __name__ == '__main__':
    app.run(debug=True, port=5000)
```
